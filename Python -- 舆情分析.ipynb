## 策略思想 

1、如果当周的谷歌Debt搜索量>过去三周平均，则做空道琼斯指数，持仓一周  #用周平均计算的
2、如果当周的谷歌Debt搜索量<过去三周平均，则做多道琼斯指数，持仓一周

```python
import pandas as pd
import numpy as np
import quandl
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
```

## 数据处理

```python
paper = pd.read_csv(
    'data/paper_data.csv',
    sep=' ',  #数据分割的方式，在CSV里的数据是用空格隔开的,这里分割的方式就是空格
    parse_dates=True)  #把时间解析为时间数据datetime
paper.head()
```

```python
data = pd.DataFrame({
    'Google_week': paper['Google End Date'],
    'debt': paper['debt'].astype(np.float64),
    'Date': paper['DJIA Date'],
    'DJClose': paper['DJIA Closing Price'].astype(np.float64)
})
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])  #转换为时间序列格式
```

```python
data['Google_week'] = pd.to_datetime(data['Google_week'])
```

## 读取自己下载得谷歌搜索指数数据

```python
trends_download = pd.read_csv('data/debt_google_trend.csv')
trends_download.head()
```

```python
trends_download['Week'] = trends_download['Week'].apply(
    lambda x: pd.to_datetime(x.split(' ')[-1]))  #将Week里面得数据用空格隔开，取最后一个日期
trends_download.head()
```

```python
all_data = pd.merge(data,
                    trends_download,
                    left_on='Google_week',
                    right_on='Week')
```

```python
all_data.head()
```

```python
all_data.drop('Week', inplace=True, axis=1)
all_data.set_index('Date', inplace=True)
```

```python
all_data.head()
```

```python
all_data.rename(columns={
    'debt_x': 'Debt_paper',
    'debt_y': 'Debt_download'
},
                inplace=True)
```

```python
all_data.head()
```

```python
# 查看自己下载得数据和论文里数据得相关性，为了验证最后结论得有效性
both_trends = all_data[['Google_week', 'Debt_paper',
                        'Debt_download']].set_index('Google_week')
both_trends.head()
```

```python
#计算相关性
both_trends.corr()
```

## 策略实现

```python
all_data = all_data.reset_index().set_index('Google_week')
all_data.head()
```

```python
all_data['MA_p'] = all_data['Debt_paper'].shift(1).rolling(3).mean()    #这里策略是在第四周得时候得debt搜索量跟前三周得均值相比，在第五周才开始交易
all_data['MA_d'] = all_data['Debt_download'].shift(1).rolling(3).mean()
all_data.head()
```

```python
all_data['signal_p'] = np.where(all_data['Debt_paper']>all_data['MA_p'], -1, 1)
all_data['signal_d'] = np.where(all_data['Debt_download']>all_data['MA_d'], -1, 1)
```

```python
all_data.head()
```

```python
all_data.loc[:3,['signal_p','signal_d']] = 0
```

```python
all_data.head()
```

## 计算策略收益并可视化

```python
all_data['pct_change'] = all_data['DJClose'].pct_change()
```

```python
all_data.head()
```

```python
all_data['ret_p'] = all_data['pct_change'] * all_data['signal_p'].shift(1)
all_data['ret_d'] = all_data['pct_change'] * all_data['signal_d'].shift(1)
```

```python
all_data.head()
```

```python
# 计算累计收益
all_data['cum_ret_p'] = (all_data['ret_p']+1).cumprod()
all_data['cum_ret_d'] = (all_data['ret_d']+1).cumprod()
```

```python
sns.set()
all_data[['cum_ret_p','cum_ret_d']].plot(figsize=(10,6))
```
